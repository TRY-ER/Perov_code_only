```python
import pandas as pd
import numpy as np
import joblib
```

```python
main_df = pd.read_csv("./outputs/data/space_preprocessed_main.csv")
# main_df = pd.read_csv("./inputs/smote_main_reversed_revised_main.csv")
```

```python
main_df.head()
```

```python
prop_dict = {}
for col in main_df.columns:
    try:
        prop_dict[col] = {}
        prop_dict[col]["min"] = main_df[col].min()
        prop_dict[col]["max"] = main_df[col].max()    
        prop_dict[col]["count"] = main_df[col].count()    
        prop_dict[col]["u_val"] = main_df[col].unique()
        prop_dict[col]["unique_count"] = main_df[col].nunique()
        prop_dict[col]["type"] = None
        try:
            if col[-4:] == "bool":
                prop_dict[col]["type"] = "bool"
                print(col)
            else:
                float(prop_dict[col]["min"])
                prop_dict[col]["type"] = "numeric"
        except:
            prop_dict[col]["type"] = "categorical"
    except:print(col)
```

```python
for key,item in prop_dict.items():
    if item["type"] == "bool":
        print(key)
```

```python
joblib.dump(prop_dict,"./deposition/col_data_pred.z")
```
