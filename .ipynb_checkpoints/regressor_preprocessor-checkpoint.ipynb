```python
from sklearn.preprocessing import LabelEncoder, StandardScaler
```

```python
for col in main_df.columns:
    if col in categorical or col in boolean:
        lbl_enc = LabelEncoder()
        main_df[col] = lbl_enc.fit_transform(main_df[col])
        joblib.dump(lbl_enc,f"./testing/label_encoders/{col}_lbl_enc.z")
```

```python
col_data_testing = {"numerical": numeric,
                    "categorical":categorical,
                    "boolean":boolean}
```

```python
joblib.dump(col_data_testing, "./testing/label_encoders/col_data.z")
```

```python
shaped_df = main_df.drop(["PCE_categorical"],axis=1)
scaler = StandardScaler()
scaled_df = scaler.fit_transform(shaped_df)
joblib.dump(scaler,"./testing/scaler.z")
```

```python
export_df = pd.DataFrame(scaled_df,columns=[col for col in main_df.columns if col != "PCE_categorical"] )
export_df["PCE_categorical"] = main_df["PCE_categorical"]
```
