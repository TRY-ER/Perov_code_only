```python
col_data = joblib.load("./propdict.z")
```

```python
encoders = os.listdir("./encoders")
```

```python
exclude_lbl_enc = lambda x : x[:-2]
enc = list(map(exclude_lbl_enc,encoders))
```

```python
# enc = joblib.load("./encoders/Substrate_stack_sequence_0_.z")
# print(enc.transform(["pin"]))
mod_df = predict_df.copy()
for col in predict_df.columns:
    if col in enc:
        lbl_enc = joblib.load(f"./encoders/{col}.z")
        mod_df[col] = lbl_enc.transform(mod_df[col])
        # print(col)
```

```python
standard_scaler = joblib.load("./scaler/std_scaler.z")
mod_df = standard_scaler.transform(mod_df)
mod_df = pd.DataFrame(mod_df,columns=predict_df.columns)
```

```python
num_folds = len(os.listdir("./models/"))
#loading the model
def predict(df,num_folds):
    result_li = []
    for fold in range(num_folds):
        print(f"predicting for fold {fold} / {num_folds}")
        model = joblib.load(f"./models/{fold}_xg_boost_reg_no_optim/{fold}_model.z")
        result = model.predict(df)
        print(result)
        result_li.append(result)
    return result_li
```

```python
result = predict(mod_df,num_folds)
value = np.mean(result)
print(f" Regressor says value should be around :: {value}")
```
