```python
import pandas as pd
import numpy as np
import joblib
```

```python
train_df = pd.read_csv("./testing/preprocessed_dataset.csv")
```

```python
from sklearn.model_selection import StratifiedKFold
from sklearn.metrics import accuracy_score
from xgboost import XGBClassifier
```

```python
model = XGBClassifier(n_estimators = 500,
                      learning_rate = 0.1,
                      booster = "dart",
                      tree_method = "gpu_hist",
                      predictor = "cpu_predictor")
```

```python
skf = StratifiedKFold(n_splits = 5, shuffle=True, random_state=123)
```

```python
acc_score = []
X = train_df.drop(["PCE_categorical"], axis=1)
Y = train_df["PCE_categorical"]
for i,(train_idx,test_idx) in enumerate(skf.split(X,Y)):
    X_train,X_test = X.iloc[train_idx,:].to_numpy(dtype=np.float64),X.iloc[test_idx,:].to_numpy(dtype=np.float64)
    Y_train,Y_test = Y.iloc[train_idx].to_numpy(dtype=np.float64),Y.iloc[test_idx].to_numpy(dtype=np.float64)
    print(f"[{i}/5] training started")
    model.fit(X_train,Y_train,
              eval_set = [(X_test,Y_test)],
              eval_metric = ["merror"])
    Y_pred = model.predict(X_test)
    joblib.dump(model,f"./testing/models/{i}_xgboost.z")
    acc = accuracy_score(Y_pred,Y_test)
    
```
